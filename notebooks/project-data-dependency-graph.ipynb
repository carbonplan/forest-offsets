```python
from IPython.display import Image
from pygraphviz import AGraph
```

```python
def project_dag():
    G = AGraph()
    G.add_node("fig_4a", label="Figure 4a")

    inputs = {
        "fia_bio": "FIA Biomass",
        "fia_tree": "FIA TREE",
        "fia_for": "FIA FORTYPCDs",
        "aa_species": "Forest Codes per Assessment Area",
        "ss": "Supersection Shapes",
        "aa_carb": "CARB Assessment Area Data File",
        "opdrs": "Project OPDRs",
        "proj_geo": "Project Shapefiles",
        "issuance": "CARB Issuance Table",
        "prism": "PRISM 30-yr Normals",
    }

    outputs = {
        "retro_db": "Forest Offset DB (json/csv)",
        "project_db": "Project DB",
        "reclass": "Reclassification",
        "rfia": "rFIA Biomass",
    }

    figures = {
        "fig_1": "Fig 1: Program Summary",
        "fig_2": "Fig 2: Baseline Convergence",
        "fig_3": "Fig 3: Crediting Error",
        # fig_4': "Fig 4: Southern Cascades"
    }

    fig_4 = {
        "fig_4a": "Fig 4a: Southern Cascades Arbitrage",
        "fig_4b": "Fig 4b: Southern Cascades by Ecosection",
    }

    for node, label in inputs.items():
        G.add_node(node, label=label, shape="box", face="grey")

    for node, label in outputs.items():
        G.add_node(node, label=label, shape="invhouse")

    for node, label in figures.items():
        G.add_node(node, label=label, shape="ellipse")

    for node, label in fig_4.items():
        G.add_node(node, label=label, shape="ellipse")

    # fig 1
    G.add_edge("issuance", "fig_1")
    G.add_edge("retro_db", "fig_1")

    # retro-json
    G.add_edge("opdrs", "project_db")
    G.add_edge("project_db", "retro_db")
    G.add_edge("proj_geo", "retro_db")

    #
    G.add_edge("fia_tree", "reclass")
    G.add_edge("fia_for", "reclass")
    G.add_edge("retro_db", "reclass")
    G.add_edge("reclass", "fig_3")
    G.add_edge("rfia", "fig_3")

    #
    G.add_edge("retro_db", "fig_2")
    G.add_edge("retro_db", "fig_3")
    G.add_edge("retro_db", "fig_4a")
    G.add_edge("retro_db", "fig_4b")

    # figure 4
    ## figure 4a
    G.add_edge("fia_bio", "fig_4a")
    G.add_edge("prism", "fig_4a")

    G.add_edge("ss", "rfia")
    G.add_edge("aa_species", "rfia")
    G.add_edge("rfia", "fig_4b")

    G.add_edge("aa_carb", "fig_4b")

    return G


Image(project_dag().draw(prog="dot", format="png"))
```
